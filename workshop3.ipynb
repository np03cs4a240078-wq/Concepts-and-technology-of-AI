```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/AI/Titanic-Dataset.csv")
df.info()
df.head()
```

```python
fare = df["Fare"]
fare.head()
```

```python
class_age = df[["Pclass", "Age"]]
class_age.head()
```

```python
survived_gender = df[["Survived", "Sex"]]
survived_gender.head()
```

```python
fare_gt_100 = df[df["Fare"] > 100]
fare_gt_100
```

```python
first_class = df[df["Pclass"] == 1]
first_class
```

```python
female_under_18 = df[(df["Age"] < 18) & (df["Sex"] == "female")]
female_under_18
```

```python
embarked_c_or_s = df[df["Embarked"].isin(["C", "S"])]
embarked_c_or_s
```

```python
first_second_class = df[df["Pclass"].isin([1,2])]
first_second_class
```

```python
missing_values = df["Age"].isnull().sum()
missing_values
```

```python
df["Age"] = df['Age'].fillna(df['Age'].median())
df.isnull().sum()
```

```python
df["fare_per_year"] = df["Fare"] / df["Age"]
df.head()
```

```python
high_fare_age = df[df["fare_per_year"] > 5]
high_fare_age
```

```python
high_fare_age_srt = high_fare_age.sort_values(by = "fare_per_year", ascending=False)
high_fare_age_srt
```

```python
result = high_fare_age_srt[["Name","fare_per_year"]]
result
```

```python
df["fare_per_class"] = df["Fare"] / df["Pclass"]
df.head()
```

```python
adult_males = df[(df['Age'] >= 18) & (df['Sex'] =="male")]
adult_males
```

```python
adult_males_srt = adult_males.sort_values(by = "fare_per_class", ascending = False)
adult_males_srt
```

```python
result = adult_males_srt[["Name", "Age", "fare_per_class"]]
result
```

```python
total_fare = df["Fare"].sum()
total_fare
```

```python
fare_total_list_by_class = (df.groupby("Pclass")["Fare"].sum()).tolist()
fare_total_list_by_class
```

```python
fare_revenue_by_class = np.array(fare_total_list_by_class) / total_fare
fare_revenue_by_class
```

```python
bins = [0, 17, 65, float("inf")]
labels = ["child", "adult", "senior"]

df["age_group"] = pd.cut(df["Age"], bins = bins, labels = labels)
df.head(10)
```

```python
total_passengers = len(df)
total_passengers
```

```python
total_passengers_by_age_group = df.groupby("age_group")["PassengerId"].count()
total_passengers_by_age_group
```

```python
proportion_by_age_group = total_passengers_by_age_group / total_passengers
proportion_by_age_group
```

```python
percentage_by_age_group = proportion_by_age_group * 100
percentage_by_age_group = percentage_by_age_group.round(2)
percentage_by_age_group
```
